# Modelo de predicion

    
   # Regresion simple


```python
w=125
b=0
```

```python
var="GrLivArea"

x=np.linspace(0,df_train[var].max(),100)
y=w*x+b

df_train.plot.scatter(x=var,y="SalePrice",ylim=(0,df_train["SalePrice"].max()*1.1))
plt.plot(x,y,"-r")
plt.show
```

```python
from sklearn.linear_model import LinearRegression

#importar input y output
X_train=np.array(df_train["GrLivArea"]).reshape((-1,1))
Y_train=np.array(df_train["SalePrice"])
                 
# crear modelo

model=LinearRegression(fit_intercept=False)
model.fit(X_train, Y_train)

#imprimir parametros

print(f"intercepto (b):{model.intercept_}")
print(f"pendiente (w):{model.coef_}")
```

```python
w=118.069
b=0
```

```python
#calculo de predicion
df_train["pred"]=df_train["GrLivArea"]*w+b
```

```python
#calculo de la funcion de error
df_train["diff"]=df_train["pred"]-df_train["SalePrice"]
df_train["cuad"]=df_train["diff"]**2
df_train.head()
```

```python
df_train["cuad"].mean()
```

```python
from sklearn.metrics import r2_score
```

```python
print("Efectividad: %2f" % (r2_score(df_train["SalePrice"],df_train["pred"])*100))
```

Con la regresion simple nos da una efectividad por debajo del 50%
Tenemos que considerar que este modelo se construyo con una sola variable para aumentar el porcentaje haremos un modelo mas robusto con las demas variables de interes. 

De no funcionar aplicariamos otra tecnica para predecir la variable "SalePrice".
